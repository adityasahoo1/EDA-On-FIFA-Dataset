```python
# Importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv(r'D:\Github\EDA on FIFA Dataset\players_20.csv')
```

```python
fifa['nationality'].value_counts()[0:10]
```

```python
# Top 10 countries with highest number of players

plt.figure(figsize=(10,5))
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()), list(fifa['nationality'].value_counts()[0:10]))
plt.xlabel("Countries")
plt.title("Top 10 countries with highest number of players")
plt.show()
```

```python
# Percentage of top 10 countries by number of players

plt.pie(list(fifa['nationality'].value_counts()[0:10]), labels = fifa['nationality'].value_counts()[0:10].keys(), radius = 2, autopct = '%1.01f%%')
plt.pie([1],radius = 1, colors = 'w')
plt.show()
```

```python
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary = player_salary.sort_values(by = 'wage_eur', ascending = False)
```

```python
player_salary.head()
```

```python
# Top 5 players with highest income

plt.figure(figsize=(9,7))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5])
plt.title("Top 5 players with highest income")
plt.xlabel("Players")
plt.show()
```
